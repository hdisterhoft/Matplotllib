```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_reader = pd.read_csv(city_data_to_load, encoding="ISO-8859-1")
ride_reader = pd.read_csv(ride_data_to_load, encoding="ISO-8859-1")
# Combine the data into a single dataset
city_ride_df = pd.merge(city_reader, ride_reader, on="city")

# Display the data table for preview
city_ride_df.head()
```

```python
urban_group = ride_reader.groupby(['city'])
urban_group2=urban_group.mean()


urban_df = city_reader[city_reader.type != 'Rural']

urban_df2 = urban_df[urban_df.type != 'Suburban']


urban_df3 = pd.merge(urban_df2, urban_group2, on="city")
urban_df3['colors']='red'


urban_df3
```

```python
suburban_group = ride_reader.groupby(['city'])
suburban_group2=suburban_group.mean()


suburban_df = city_reader[city_reader.type != 'Rural']

suburban_df2 = suburban_df[suburban_df.type != 'Urban']


suburban_df3 = pd.merge(suburban_df2, suburban_group2, on="city")

suburban_df3['colors']='lightskyblue'
suburban_df3
```

```python
rural_group = ride_reader.groupby(['city'])
rural_group2=rural_group.mean()


rural_df = city_reader[city_reader.type != 'Suburban']

rural_df2 = rural_df[rural_df.type != 'Urban']


rural_df3 = pd.merge(rural_df2, rural_group2, on="city")

rural_df3['colors']='yellow'
rural_df3
```

```python
frames = [urban_df3,suburban_df3,rural_df3]
final_df = pd.concat(frames)
final_df
```
